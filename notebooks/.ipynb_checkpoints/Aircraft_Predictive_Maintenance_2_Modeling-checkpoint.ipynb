# <div style="text-align: left; background-color: yellowgreen; color: white; padding: 10px; line-height:1;border-radius:10px">2. Data Loading and Preparation</div>

In this code section, we implement crucial data preprocessing steps, with a particular focus on addressing class imbalance - a common challenge in predictive maintenance applications. The code performs several essential operations: data loading, feature standardization, and most importantly, class imbalance correction through two complementary techniques: class weights and SMOTE (Synthetic Minority Over-sampling Technique).

The need for these balancing techniques stems from the inherent nature of maintenance data. 

In real-world scenarios: 
- equipment typically spends most of its time in normal operating conditions (class 0), 
- with relatively few instances of warning states (class 1) 
- or critical conditions (class 2). 

This natural imbalance can severely impact model performance if not properly addressed.

Without proper balancing:
- The model would be biased towards predicting the majority class (normal condition)
- It would likely miss critical warning signs of impending failures
- The cost of missing these warning signs (false negatives) is typically much higher than false positives

The implementation uses two complementary approaches: Class Weights & SMOTE

The role of class weights is to assigns higher importance to minority classes during model training and to penalize misclassification of minority classes more heavily.

The role of SMOTE is to create synthetic examples of minority classes using nearest neighbors and to help the model better learn the decision boundaries for underrepresented classes.

```python
def load_data():
    """
    Load and prepare training and test datasets.
    
    Returns:
        tuple: Processed datasets (X_train, X_test, y_train, y_test, features)
            - X_train, X_test: Standardized feature matrices
            - y_train, y_test: Target vectors
            - features: List of feature names
            - class_weight_dict: Dictionary of class weights
    """
    # Load datasets
    df_train = pd.read_csv('../data/train.csv')
    df_test = pd.read_csv('../data/test.csv')
    
    # Define features
    features = ["setting1", "setting2", "setting3", 
               "s1", "s2", "s3", "s4", "s5", "s6", "s7", "s8", "s9", "s10", 
               "s11", "s12", "s13", "s14", "s15", "s16", "s17", "s18", "s19", 
               "s20", "s21"]
    
    # Split features and target
    X_train = df_train[features]
    X_test = df_test[features]
    y_train = df_train['label_mcc']
    y_test = df_test['label_mcc']
    
    # Standardize features
    scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)
    
    # Add class weights calculation
    class_weights = compute_class_weight(
        class_weight='balanced',
        classes=np.unique(y_train),
        y=y_train
    )
    class_weight_dict = dict(zip(np.unique(y_train), class_weights))
    
    print("Class distribution before balancing:")
    print(pd.Series(y_train).value_counts().sort_index())
    
    # Apply SMOTE to balance classes
    smote = SMOTE(random_state=42)
    X_train_balanced, y_train_balanced = smote.fit_resample(X_train, y_train)
    
    print("\nClass distribution after balancing:")
    print(pd.Series(y_train_balanced).value_counts().sort_index())
    
    return X_train_balanced, X_test, y_train_balanced, y_test, features, class_weight_dict
```

```python
X_train, X_test, y_train, y_test, features, class_weight_dict = load_data()
```

After applying SMOTE, we achieve a balanced distribution with 17,531 instances per class. 

This balanced dataset provides several advantages:
- Equal representation of all operational states
- Better learning of warning signs patterns
- Improved model sensitivity to critical conditions
- More robust decision boundaries between classes

# <div style="text-align: left; background-color: yellowgreen; color: white; padding: 10px; line-height:1;border-radius:10px">3. Model Training</div>

## 3.1. The definition of a custom score

In predictive maintenance, the timing of maintenance interventions is crucial. While traditional scoring metrics like accuracy treat all misclassifications equally, our specific use case requires a more nuanced approach. The custom scoring function is designed to optimize the model's ability to identify the "sweet spot" for maintenance interventions, represented by Class 1.

Class 1 represents an optimal intervention window where:
- The equipment shows early signs of degradation
- There's enough time to plan and schedule maintenance
- The risk of immediate failure is still manageable
- Maintenance costs are typically lower than emergency repairs

```python
def custom_scoring(y_true, y_pred):
    """
    Custom scoring function that penalizes misclassification of class 1 as class 0.
    
    Args:
        y_true: True labels
        y_pred: Predicted labels
        
    Returns:
        float: Score where misclassifying class 1 as 0 is heavily penalized
    """
    # Claculation of a confusion matrix to define a specific "recall custom score"
    cm = confusion_matrix(y_true, y_pred)
    # Specific penalty for classifying class 1 as class 0
    false_negatives_1_as_0 = cm[1][0] 
    
    # Calculate standard metrics
    accuracy = metrics.accuracy_score(y_true, y_pred)
    recall_class_1 = metrics.recall_score(y_true, y_pred, labels=[1], average=None)[0]
    
    # Custom score that heavily weights recall for class 1
    score = (0.7 * recall_class_1 + 0.3 * accuracy) * (1 - 0.1 * false_negatives_1_as_0)
    return score
```

The scoring function implements this business logic through three components:
- Recall Weight (0.7)
- Accuracy Component (0.3)
- Penalty Factor

This scoring approach ensures that:
- The model is optimized for detecting maintenance opportunities
- Critical misclassifications are heavily penalized
- Overall accuracy is maintained as a secondary objective
- The training process aligns with real-world maintenance priorities

By using this custom scoring function during model selection and evaluation, we create a predictive maintenance system that better reflects the practical requirements of maintenance planning and cost optimization.

## 3.2. The training properly said

The training strategy implements a Random Forest classifier specifically tuned for predictive maintenance.

Hyperparameter Optimization Strategy is managed by GridSearchCV. GridSearchCV will explore the dictionnary of parameters predefined as "params", in order to find the best hyperparameters.  

To ensure reliable result, a Cross-Validation process is performed with a 5-fold cross-validation with ShuffleSplit, with validation split size of 25%. 

This optimization process is performed specifically with the custom metrics defined previously as target.

```python
def train_model(X_train, y_train, class_weight_dict):
    """
    Train a Random Forest model with hyperparameter optimization.
    
    Args:
        X_train: Training feature matrix
        y_train: Training target vector
        class_weight_dict: Dictionary of class weights
    
    Returns:
        GridSearchCV: Trained model with best parameters
    """
    # Define model and parameters grid
    clf = RandomForestClassifier(
        n_estimators=100,
        random_state=21,
        class_weight=class_weight_dict,
        min_samples_leaf=2,
        max_features='sqrt'
    )
    params = {
        'max_depth': [4, 5, 6, 7, 8],
        'criterion': ['gini', 'entropy'],
        'min_samples_split': [2, 5, 10],
        'class_weight': [class_weight_dict, 'balanced', 'balanced_subsample']
    }
    
    # Cross-validation strategy
    ss = ShuffleSplit(n_splits=5, test_size=0.25, random_state=21)
    
    # Custom scorer that penalizes false negatives for class 1
    custom_scorer = make_scorer(custom_scoring)
    
    # Grid search with cross-validation
    grid_search = GridSearchCV(
        estimator=clf,
        param_grid=params,
        cv=5,
        scoring=custom_scorer,
        n_jobs=-1,
        verbose=1
    )
    
    grid_search.fit(X_train, y_train)
    return grid_search
```

```python
model = train_model(X_train, y_train, class_weight_dict)
```

# <div style="text-align: left; background-color: yellowgreen; color: white; padding: 10px; line-height:1;border-radius:10px">4. Model Evaluation</div>

The evaluation function implements a thorough assessment strategy that particularly emphasizes the model's ability to detect the optimal maintenance window (Class 1 - Low Risk). This comprehensive evaluation is crucial for predictive maintenance applications where missing intervention opportunities can lead to significant consequences.

Recall: Measures our ability to detect all maintenance opportunities

Precision: Indicates how reliable our maintenance alerts are

```python
def evaluate_model(model, X_test, y_test):
    """
    Evaluate model performance and display metrics.
    
    Args:
        model: Trained model
        X_test: Test feature matrix
        y_test: Test target vector
    
    Returns:
        tuple: (predictions, probability scores)
    """
    # Get predictions
    y_pred = model.predict(X_test)
    y_score = model.predict_proba(X_test)
    
    # Calculate and display class-specific metrics
    print("\nDetailed Classification Report:")
    print(classification_report(y_test, y_pred, target_names=['Class 0', 'Class 1', 'Class 2']))
    
    # Calculate specific metrics for class 1
    class_1_recall = metrics.recall_score(y_test, y_pred, labels=[1], average=None)[0]
    class_1_precision = metrics.precision_score(y_test, y_pred, labels=[1], average=None)[0]
    
    print("\nClass 1 Specific Metrics:")
    print(f"Recall (Class 1): {class_1_recall:.3f}")
    print(f"Precision (Class 1): {class_1_precision:.3f}")
    
    # Create confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    
    # Display confusion matrix with absolute values
    plt.figure(figsize=(10, 7))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=['Normal', 'Low Risk', 'High Risk'],
                yticklabels=['Normal', 'Low Risk', 'High Risk'])
    plt.title('Confusion Matrix (Absolute Values)', pad=20)
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    
    # Add total counts per class
    plt.text(-0.4, -0.1, f'Total actual Normal: {sum(cm[0])}', fontsize=10)
    plt.text(-0.4, -0.2, f'Total actual Low Risk: {sum(cm[1])}', fontsize=10)
    plt.text(-0.4, -0.3, f'Total actual High Risk: {sum(cm[2])}', fontsize=10)
    
    plt.tight_layout()
    plt.show()
    
    # Display critical misclassification details
    print("\nMisclassification Analysis:")
    print(f"Low Risk cases predicted as Normal: {cm[1][0]} out of {sum(cm[1])} Low Risk cases")
    print(f"High Risk cases predicted as Normal: {cm[2][0]} out of {sum(cm[2])} High Risk cases")
    
    # Display accuracy per class
    print("\nAccuracy per class:")
    print(f"Normal cases correctly predicted: {cm[0][0]} out of {sum(cm[0])} ({cm[0][0]/sum(cm[0]):.1%})")
    print(f"Low Risk cases correctly predicted: {cm[1][1]} out of {sum(cm[1])} ({cm[1][1]/sum(cm[1]):.1%})")
    print(f"High Risk cases correctly predicted: {cm[2][2]} out of {sum(cm[2])} ({cm[2][2]/sum(cm[2]):.1%})")
    
    return y_pred, y_score
```

```python
y_pred, y_score = evaluate_model(model, X_test, y_test)
```

The predictive maintenance model demonstrates strong performance in its critical objective: detecting maintenance opportunities. 

With a recall of 87% for the maintenance window (Class 1) and, most importantly, zero cases of missed maintenance opportunities (no Low Risk cases classified as Normal), the model achieves its primary safety goal. 

While this leads to some over-cautious predictions (19 Normal cases identified as Low Risk), this trade-off is acceptable in a maintenance context where false alarms are preferable to missed failures. The model also performs well in detecting High Risk situations, with no critical cases being misclassified as Normal operation. 

Overall accuracy of 76% indicates reliable general performance while successfully prioritizing the detection of maintenance windows.

## 5.1. Feature importance

Feature importance analysis identifies and ranks the sensors that have the strongest influence on the model's predictions for equipment failure risk.

```python
FEATURE_MAPPING = {
    'setting1': 'setting_1',
    'setting2': 'setting_2',
    'setting3': 'setting_3',
    's1': '(Fan inlet temperature) (◦R)',
    's2': '(LPC outlet temperature) (◦R)',
    's3': '(HPC outlet temperature) (◦R)',
    's4': '(LPT outlet temperature) (◦R)',
    's5': '(Fan inlet Pressure) (psia)',
    's6': '(bypass-duct pressure) (psia)',
    's7': '(HPC outlet pressure) (psia)',
    's8': '(Physical fan speed) (rpm)',
    's9': '(Physical core speed) (rpm)',
    's10': '(Engine pressure ratio(P50/P2)',
    's11': '(HPC outlet Static pressure) (psia)',
    's12': '(Ratio of fuel flow to Ps30) (pps/psia)',
    's13': '(Corrected fan speed) (rpm)',
    's14': '(Corrected core speed) (rpm)',
    's15': '(Bypass Ratio) ',
    's16': '(Burner fuel-air ratio)',
    's17': '(Bleed Enthalpy)',
    's18': '(Required fan speed)',
    's19': '(Required fan conversion speed)',
    's20': '(High-pressure turbines Cool air flow)',
    's21': '(Low-pressure turbines Cool air flow)'
}
```

```python
def plot_feature_importance(model, features):
    """
    Visualize feature importance from the trained model.
    
    Args:
        model: Trained model with feature_importances_ attribute
        features: List of feature names
    """
    # Get feature importances
    importances = model.best_estimator_.feature_importances_
    
    # Create a DataFrame for easier manipulation
    importance_df = pd.DataFrame({
        'feature': features,
        'importance': importances
    })
    
    # Create combined labels with both short and full names
    importance_df['combined_label'] = importance_df['feature'].apply(
        lambda x: f"{x} - {FEATURE_MAPPING[x]}" if x in FEATURE_MAPPING else x
    )
    
    # Sort by importance
    importance_df = importance_df.sort_values('importance', ascending=True)
    
    # Create visualization
    plt.figure(figsize=(14, 10))
    plt.barh(range(len(importance_df)), importance_df['importance'], 
            color='g', align='center', height=0.6)
    
    # Add feature names with both short and full descriptions
    plt.yticks(range(len(importance_df)), importance_df['combined_label'], 
              fontsize=9)
    
    # Customize the plot
    plt.title('Sensor Importance in Failure Prediction', fontsize=12, pad=20)
    plt.xlabel('Relative Importance', fontsize=10)
    plt.grid(True, axis='x', linestyle='--', alpha=0.7)
    
    # Adjust layout
    plt.tight_layout()
    plt.show()
    
    # Print numerical values for top features
    print("\nTop 5 Most Important Sensors:")
    top_5 = importance_df.tail(5)[::-1]
    for _, row in top_5.iterrows():
        print(f"{row['combined_label']}: {row['importance']:.4f}")
```

```python
plot_feature_importance(model, features)
```

The feature importance analysis reveals that the model's predictions are primarily driven by five key sensors, accounting for 65% of its decision-making power. The model particularly relies on pressure measurements from the High-Pressure Compressor (HPC outlet static and dynamic pressures), complemented by the fuel flow ratio and turbine temperature readings. These findings align with engineering principles, as these parameters are fundamental indicators of engine health and performance. Notably, most other sensors have relatively low importance (below 5%), suggesting that a simplified monitoring system focusing on these top parameters could effectively detect potential failures.

## 5.2. Explainability analysis

SHAP (SHapley Additive exPlanations) values measure how each feature contributes to pushing the model's prediction higher or lower from the base prediction value, based on game theory principles. They provide transparency in machine learning models by quantifying the impact of each feature on individual predictions and across all predictions.

```python
def analyze_shap_values_summary(model, X_train, features):
    """
    Analyze and visualize SHAP values for model explanability.
    
    Args:
        model: Trained RandomForest model
        X_train: Training feature matrix
        features: List of feature names
        
    Returns:
        explainer: SHAP explainer object
        shap_values: Calculated SHAP values
    """
    # Create the SHAP explainer
    explainer = shap.TreeExplainer(model.best_estimator_)
    
    # Calculate SHAP values
    shap_values = explainer.shap_values(X_train)
    
    # Create DataFrame with full feature names
    feature_names = [f"{feat} - {FEATURE_MAPPING[feat]}" for feat in features]
    
    # Summary Plot
    shap.summary_plot(shap_values, X_train, feature_names=feature_names, 
                     plot_type="bar", class_names=['Normal', 'Low Risk', 'High Risk'])
    plt.title("Feature Impact on Model Predictions (SHAP Values)")
    plt.tight_layout()
    plt.show()
    
    return explainer, shap_values
```

```python
def analyze_shap_values_class_specific(model, X_train, features):
    """
    Analyze and visualize SHAP values for model explanability.
    
    Args:
        model: Trained RandomForest model
        X_train: Training feature matrix
        features: List of feature names
        
    Returns:
        explainer: SHAP explainer object
        shap_values: Calculated SHAP values
    """
    # Create the SHAP explainer
    explainer = shap.TreeExplainer(model.best_estimator_)
    
    # Calculate SHAP values
    shap_values = explainer.shap_values(X_train)
    
    # Create DataFrame with full feature names
    feature_names = [f"{feat} - {FEATURE_MAPPING[feat]}" for feat in features]
    
    # Class-specific Summary Plots
    for i, class_name in enumerate(['Low Risk']): #['Normal', 'Low Risk', 'High Risk']):
        plt.figure(figsize=(12, 8))
        shap.summary_plot(shap_values[i], X_train, feature_names=feature_names,
                         plot_type="violin", show=False)
        plt.title(f"SHAP Values Distribution for {class_name} Prediction")
        plt.tight_layout()
        plt.show()
    
    return explainer, shap_values
```

```python
print("Feature Impact on Model Predictions (SHAP Values)")
explainer, shap_values = analyze_shap_values_summary(model, X_train, features)
```

The SHAP Summary chart displays the absolute magnitude of feature impacts across all risk levels (Normal, Low Risk, High Risk)
- It confirms s11 (HPC outlet Static pressure) as the dominant predictor across all risk classes
- It shows how different features contribute differently to each risk level, with some features (like s4 and s7) being particularly important for High Risk predictions
- Most operational settings (bottom of graph) have minimal impact on predictions

```python
print("Class Specific Analysis")
explainer, shap_values = analyze_shap_values_class_specific(model, X_train, features)
```

The SHAP Values distribution for class 1 explain the main feature implied in class 1 predictions. It shows how each sensor impacts Low Risk predictions through violin plots
- The HPC outlet static pressure (s11) has the widest distribution, indicating strong but variable impact
- Core engine parameters (s4, s7, s12) show distinct bi-modal distributions, suggesting they behave differently in different operational contexts

```python
def save_model_joblib(model, filename):
    """
    Save the trained model to disk.
    
    Args:
        model: Trained model to save
        filename: Output filename for the model
    """
    dump(model, filename)
    print(f"Model saved as {filename}")

def save_model_pickle(model, filename):
    """
    Save the trained model to disk.
    
    Args:
        model: Trained model to save
        filename: Output filename for the model
    """
    with open(filename, 'wb') as file:
        pickle.dump(model, file)
    print(f"Model saved as {filename}")
    
save_model_joblib(model, '../models/pred_maint_random_forest_2024_11_01.joblib')
```
