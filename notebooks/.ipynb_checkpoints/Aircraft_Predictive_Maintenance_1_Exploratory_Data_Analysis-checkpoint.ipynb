## 2.1 Dataset description

### Data Source

The dataset can be downloaded on Kaggle.com at the following [weblink](https://www.kaggle.com/datasets/mhadani/predictive-maintenance-aircraft-engine)

This dataset is designed for predictive maintenance of aircraft engines and consists of three primary files:

Training Data ("PM_train.csv"): This file contains multiple multivariate time series data, with each time series representing the operational cycles of different aircraft engines of the same type. Each cycle includes 21 sensor readings. Engines start with varying initial wear and manufacturing differences, which are unknown to the user. Initially, engines operate normally and begin to degrade over time. The degradation increases until a predefined threshold is reached, marking the engine as unsafe for further use. The final cycle in each time series indicates the failure point of the engine.

Testing Data ("PM_test.csv"): This file shares the same schema as the training data but does not specify the failure points. For example, an engine might run from cycle 1 to cycle 31 without indicating how many more cycles it can last before failure.

Ground Truth Data ("PM_truth.csv"): This file provides the actual remaining working cycles for the engines in the testing data. For instance, it shows that an engine running from cycle 1 to cycle 31 in the testing data has 112 remaining cycles before failure.

# <div style="text-align: left; background-color: yellowgreen; color: white; padding: 10px; line-height:1;border-radius:10px">3. Exploratory Data Analysis</div>

Exploratory Data Analysis (EDA) is an approach in data analysis that consists of examining and understanding data before moving on to more complex analyses or applying sophisticated statistical models.
It's a crucial preliminary step that allows you to:

- Understand data structure and quality (missing values, errors, variable types)
- Identify main trends and relationships between variables
- Detect anomalies and outliers
- Generate hypotheses to explore further

This exploration is typically done through simple descriptive statistics (mean, median, standard deviation) and visualizations (histograms, box plots, scatter plots).
The importance of EDA lies in its ability to prevent interpretation errors later in the analysis and guide the choice of statistical methods to use. It's like surveying a plot of land before building: you need to know the nature of the soil before choosing the appropriate foundation.
This step is considered fundamental by data scientists because it influences the quality and relevance of all subsequent analyses.

A correlation map (or correlation matrix) is a visual representation that shows the strength and direction of relationships between multiple variables in a dataset, typically displayed as a color-coded matrix where each cell represents the correlation coefficient between two variables. 
The correlation coefficients range from -1 to +1, where -1 indicates a perfect negative correlation (as one variable increases, the other decreases), +1 indicates a perfect positive correlation (both variables increase together), and 0 means no correlation.

# <div style="text-align: left; background-color: yellowgreen; color: white; padding: 10px; line-height:1;border-radius:10px">4. A specific Focus on PCA - Principal Component Analysis</div>

Principal Component Analysis (PCA) is a dimensionality reduction technique that transforms a dataset with potentially correlated features into a smaller set of uncorrelated components, called principal components, which capture the most variance in the data. It works by finding directions (principal components) along which the variance in the data is maximized, allowing for a more compact representation. PCA is often used to simplify datasets, visualize high-dimensional data, or improve the efficiency of machine learning algorithms.

```python
train = pd.read_csv('train.csv')
```

```python
col_to_analyze = ['setting1','setting2','setting3','s1','s2','s3','s4','s5','s6','s7','s8','s9','s10','s11',
                  's12','s13','s14','s15','s16','s17','s18','s19','s20','s21']
```

```python
df_PCA = train[col_to_analyze]
target = train['label_mcc']
target2 = train['ttf']
```

There are a couple of requirements to perform a PCA analysis : 

- PCA can not be applied if there are missing values
- PCA implies to have normalized value
- PCA implies to have numerical values

We will try a show the evolution of the variance on the 8 main components. 

```python
# Normlization of the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_PCA)
```

```python
# PCA instanciation on 8 components
n_components = 8
pca = PCA(n_components=n_components)
pca.fit(X_scaled)
```

```python
# Cumulative variance calculation
scree = (pca.explained_variance_ratio_*100).round(2)
scree_cum = scree.cumsum().round()
axe_x = np.arange(1, len(scree_cum)+1)
```

```python
plt.figure(figsize=(10,6))
plt.bar(axe_x, scree, color='steelblue')
plt.plot(axe_x, scree_cum, c='red',marker='o')
plt.xlabel("Number of components retained")
plt.ylabel("Explained Variance (%)")
plt.title("Scree Plot - Optimal Component Number Detection")
plt.xticks(np.arange(1,9,1))
plt.show(block=False)
plt.show
```

The first component explains the largest portion of the variance, around 60%, as shown by the first tall bar.
As additional components are added, each one explains less variance individually, but cumulatively (red line) the explained variance increases.
The goal is to find an "elbow point," where the cumulative explained variance levels off, indicating that additional components add minimal new information.
In this chart, the "elbow" appears to be around the third or fourth component, as adding more components after this point results in only a slight increase in explained variance. Thus, retaining around three to four components might be sufficient to capture most of the dataset's variability without unnecessary complexity.

```python
# Heatmap about the different PCA components
pcs = pca.components_
x_axis_labels=['Component_1', 'Component_2', 'Component_3', 'Component_4', 'Component_5', 'Component_6', 
               'Component_7', 'Component_8']
fig, ax = plt.subplots(figsize=(20, 6))
sns.heatmap(pcs.T, vmin=-1, vmax=1, annot=True, cmap="coolwarm", fmt="0.2f", yticklabels=col_to_analyze, 
            xticklabels=x_axis_labels)
plt.title('Heatmap showing the composition of the different components of the PCA')
```

```python
X_proj = pca.transform(X_scaled)
```

Now we can draw the distribution of the observations according to the first two components. This chart can easily be enriched with the category of risk or with the TTF indicator.

```python
# Define parameters
palette_colors = ["green", "orange", "red"]
legend_labels = {0: "Normal condition", 1: "Low risk of failure", 2: "High risk of failure"}
target_mapped = [legend_labels[t] for t in target]

# Create the subplot
fig, axes = plt.subplots(1, 2, figsize=(16, 8))

# First plot: PCA projection on 2 components - By risk category
sns.scatterplot(
    x=X_proj[:, 0], 
    y=X_proj[:, 1], 
    alpha=0.6, 
    hue=target_mapped, 
    palette=palette_colors, 
    ax=axes[0]
)
axes[0].set_title("PCA projection on 2 components - By risk category")
axes[0].legend(title="Risk Category", loc="upper right")

# Second plot: PCA projection on 2 components - with TTF gradient
scatter = sns.scatterplot(
    x=X_proj[:, 0], 
    y=X_proj[:, 1], 
    alpha=0.6, 
    hue=target2, 
    palette='magma', 
    ax=axes[1]
)

# Normalize and create color bar for second plot
norm = mcolors.Normalize(vmin=target2.min(), vmax=target2.max())
sm = plt.cm.ScalarMappable(cmap="magma", norm=norm)
sm.set_array([])
cbar = plt.colorbar(sm, ax=axes[1])
cbar.set_label("TTF (Time to Failure) Gradient")

axes[1].set_title("PCA projection on 2 components - with TTF gradient")
axes[1].legend([],[], frameon=False)  # Remove duplicate legend

plt.tight_layout()
plt.show()
```

It is interesting to see that the distribution of the points is not randomized at all. Visually, we can see that the healthy equipements are grouped together at the left of the diagram.

# <div style="text-align: left; background-color: yellowgreen; color: white; padding: 10px; line-height:1;border-radius:10px">5. PCA gif creator</div>

A snippet of code to create moving 3D plot in ".GIF" format of the PCA described above

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import imageio
import os

def create_pca_3d_gif(X_proj, target, output_filename='pca_3d_rotation.gif', n_frames=72):
    """
    Crée un GIF animé montrant une rotation 3D des données PCA avec catégories de risque
    
    Parameters:
    -----------
    X_proj : array-like
        Données projetées par PCA (utiliser les 3 premières composantes)
    target : array-like
        Labels des données pour la coloration
    output_filename : str
        Nom du fichier GIF de sortie
    n_frames : int
        Nombre d'images pour l'animation (contrôle la fluidité)
    """
    
    # Création du dossier temporaire pour les images
    if not os.path.exists('temp_frames'):
        os.makedirs('temp_frames')
    
    # Définition des couleurs et labels pour chaque classe
    colors = ['green', 'orange', 'red']
    legend_labels = {0: "Normal condition", 1: "Low risk of failure", 2: "High risk of failure"}
    
    # Création des frames pour différents angles
    for angle in range(0, 360, int(360/n_frames)):
        fig = plt.figure(figsize=(12, 10), facecolor='white')
        ax = fig.add_subplot(111, projection='3d')
        
        # Configuration du fond blanc avec grille grise
        ax.set_facecolor('white')
        
        # Suppression des panneaux de fond
        ax.w_xaxis.set_pane_color((1.0, 1.0, 1.0, 1.0))
        ax.w_yaxis.set_pane_color((1.0, 1.0, 1.0, 1.0))
        ax.w_zaxis.set_pane_color((1.0, 1.0, 1.0, 1.0))
        
        # Configuration de la grille
        ax.grid(True, color='grey', linestyle='--', alpha=0.3)
        
        # Tracé des points pour chaque classe
        for i, label in enumerate(np.unique(target)):
            mask = target == label
            ax.scatter(X_proj[mask, 0], 
                      X_proj[mask, 1], 
                      X_proj[mask, 2],
                      c=colors[i], 
                      alpha=0.5,
                      label=legend_labels[i])
        
        # Paramétrage de la vue
        ax.view_init(elev=20, azim=angle)
        ax.set_xlabel('Principal Component 1')
        ax.set_ylabel('Principal Component 2')
        ax.set_zlabel('Principal Component 3')
        ax.set_title('PCA projection on 3 components - By risk category', pad=20)
        
        # Ajout de la légende avec fond blanc
        legend = plt.legend(title="Risk Category", bbox_to_anchor=(1.15, 0.5), 
                          loc='center left', frameon=True)
        legend.get_frame().set_facecolor('white')
        legend.get_frame().set_alpha(1)
        
        # Ajustement de la mise en page
        plt.tight_layout()
        
        # Sauvegarde de l'image avec fond blanc
        filename = f'temp_frames/frame_{angle}.png'
        plt.savefig(filename, dpi=100, bbox_inches='tight', 
                   facecolor='white', edgecolor='none')
        plt.close()
    
    # Création du GIF
    frames = []
    for angle in range(0, 360, int(360/n_frames)):
        filename = f'temp_frames/frame_{angle}.png'
        frames.append(imageio.imread(filename))
    
    # Sauvegarde du GIF
    imageio.mimsave(output_filename, frames, duration=0.1)
    
    # Nettoyage des fichiers temporaires
    for angle in range(0, 360, int(360/n_frames)):
        os.remove(f'temp_frames/frame_{angle}.png')
    os.rmdir('temp_frames')

# Utilisation de la fonction avec vos données
# Ne prenez que les 3 premières composantes pour la visualisation 3D
X_proj_3d = X_proj[:, :3]

# Création du GIF
create_pca_3d_gif(X_proj_3d, target, 'pca_3d_rotation.gif')
```
